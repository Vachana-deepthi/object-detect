```python
import os
from IPython.display import Image, display

def display_images_in_folder(folder_path, width=600):
    # Get a list of all files in the folder
    file_list = os.listdir(folder_path)

    # Filter out only the image files (you may need to adjust the file extensions)
    image_files = [file for file in file_list if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))]

    # Display each image
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        display(Image(filename=image_path, width=width))

# Specify the folder path
exp2_folder_path = 'runs/detect/exp2/'

# Display images in the exp2 folder
display_images_in_folder(exp2_folder_path)
```
